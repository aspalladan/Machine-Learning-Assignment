# Pumpkin Varieties and Color
## Load up required libraries and dataset. Convert the data to a dataframe containing a subset of the data:
### Let's look at the relationship between color and variety

```python


# Convert the data to a dataframe containing a subset of the data
new_columns = ['Color', 'Origin', 'Item Size', 'Variety', 'City Name', 'Package']
new_pumpkins = pumpkins[new_columns].dropna()

# Encode categorical variables using LabelEncoder
from sklearn.preprocessing import LabelEncoder
new_pumpkins_encoded = new_pumpkins.apply(LabelEncoder().fit_transform)

```

```python
# Visualize pairplot with categorical hue
sns.pairplot(new_pumpkins_encoded, hue='Color')
plt.show()
```

```python
# Convert the data to a dataframe containing a subset of the data
new_columns = ['Color', 'Variety']
new_pumpkins = pumpkins[new_columns].dropna()
```

```python
new_pumpkins["Color"].unique()
```

```python
new_pumpkins["Variety"].unique()
```

```python


# Plotting the relationship between color and variety
plt.figure(figsize=(10, 6))
sns.countplot(x='Variety', hue='Color', data=new_pumpkins)
plt.title('Relationship between Color and Variety of Pumpkins')
plt.xlabel('Variety')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Color')
plt.tight_layout()
plt.show()
```
